```python
import pandas as pd
```

```python
ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')
ratings = pd.merge(movies,ratings).drop(['genres','timestamp'],axis=1)
ratings.head()
```

```python
user_ratings = ratings.pivot_table(index=['userId'],columns=['title'],values='rating')
user_ratings.head()
```

```python
user_ratings = user_ratings.dropna(thresh=10,axis=1).fillna(0)
user_ratings.head()
```

```python
item_similarity_df = user_ratings.corr(method='pearson')
item_similarity_df.head(100)
```

```python
# User's favorite movies and their ratings
def get_similar(movie_name, rating):
    if movie_name not in item_similarity_df.columns:
        print(f"⚠️ Movie '{movie_name}' not found in similarity matrix!")
        return pd.DataFrame(columns=["Movie", "Similarity Score"])  # Return empty DataFrame

    similar_score = item_similarity_df[movie_name] * (rating - 2.5)  # Adjust similarity based on rating
    return similar_score.reset_index().rename(columns={"title": "Movie", movie_name: "Similarity Score"})

action_lover = [("Along Came Polly (2004)", 5), ("Almost Famous (2000)", 4), ("(500) Days of Summer (2009)", 3)]

# Initialize an empty DataFrame
similar_scores = pd.DataFrame()

# Loop through each movie and fetch similar ones
for movie, rating in action_lover:
    print(f"🔍 Fetching similar movies for: {movie}")
    result = get_similar(movie, rating)

    if result.empty:
        print(f"⚠️ No similar movies found for: {movie}")
        continue  # Skip empty results

    similar_scores = pd.concat([similar_scores, result], ignore_index=True)

# **Sum similarity scores to get final recommendations**
final_recommendations = similar_scores.groupby("Movie").sum()

# **Sort by highest similarity**
final_recommendations = final_recommendations.sort_values(by="Similarity Score", ascending=False)

# **Remove already rated movies**
rated_movies = [movie for movie, _ in action_lover]
final_recommendations = final_recommendations[~final_recommendations.index.isin(rated_movies)]

# **Display the top recommended movies**
print("\n🔝 Collaborative Filtering-Based Recommended Movies:\n")
print(final_recommendations.head(20))
```
